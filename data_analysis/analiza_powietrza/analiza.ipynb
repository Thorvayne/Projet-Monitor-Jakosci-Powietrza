```python
import pandas as pd

# 1. Wczytujemy Twoje sztuczne dane (zwróć uwagę na średnik!)
df = pd.read_csv('dane_testowe.csv', sep=';')

# 2. Wyświetlamy jak to wygląda przed obróbką
print("--- SUROWE DANE ---")
print(df)

# 3. Magia statystyki: grupujemy po mieście i liczymy średnią zanieczyszczeń
srednie = df.groupby('city')[['PM10', 'PM2.5']].mean().reset_index()

# 4. Wyświetlamy wynik
print("\n--- ŚREDNIE ZAPYLENIE ---")
print(srednie)

# 5. Zapisujemy wynik obróbki do nowego pliku CSV
srednie.to_csv('gotowe_wyniki_srednie.csv', sep=';', index=False)
print("\nZrobione! Plik gotowe_wyniki_srednie.csv został utworzony.")
```
